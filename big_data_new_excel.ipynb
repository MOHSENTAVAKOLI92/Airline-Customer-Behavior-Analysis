```python
#1n
import spacy
import re
# Load the spaCy model
nlp = spacy.load("en_core_web_sm")
# Mapping of month indices to abbreviations
month_mapping = {
    1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun",
    7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"
}
# Function to extract months from the DATE entities including numeric dates
def extract_months(text):
    doc = nlp(text)
    months = []

    # First, extract using spaCy's NER
    for ent in doc.ents:
        if ent.label_ == "DATE":
            # Check for month names in the entity
            for month, full_month in month_mapping.items():
                if full_month in ent.text:
                    # Extract year if present
                    year_match = re.search(r'\b(\d{2,4})\b', ent.text)
                    year = year_match.group(0) if year_match else ''
                    months.append(f"{month_mapping[month]}-{year[-2:]}" if year else month_mapping[month])

    # Additionally, look for numeric date formats like '23/08/12'
    numeric_dates = re.finditer(r'\b(\d{1,2})/(\d{1,2})/(\d{2,4})\b', text)
    for date in numeric_dates:
        day, month, year = date.groups()
        month_idx = int(month)
        if month_idx in month_mapping:
            months.append(f"{month_mapping[month_idx]}-{year[-2:]}")

    return months

# Example usage with a DataFrame

# Apply the function to extract months
df5['Extracted_Month'] = df5['Review_content'].apply(extract_months)

print(df5[['Review_content', 'Extracted_Month']])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Add 'Successful_Booking' column based on 'Rating'
df['Successful_Booking'] = df['Rating'].apply(lambda x: 1 if x > 6 else 0)

# Convert categorical features to one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Traveller_type', 'Class', 'Route', 'Flying_month'])

# Define features (X) and target (y)
X = df_encoded.drop(columns=['Rating', 'Review_title', 'Review_content', 'Successful_Booking'])
y = df_encoded['Successful_Booking']

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'ROC-AUC: {roc_auc}')

# Feature importance
feature_importance = pd.DataFrame(model.feature_importances_, index=X.columns,
 columns=['importance']).sort_values('importance', ascending=False)

# Plot the top 20 important features with brighter colors
plt.figure(figsize=(8, 4))
sns.barplot(x='importance', y=feature_importance.index[:20],
            data=feature_importance[:20], palette='Set2')
plt.title('Top 20 Feature Importances')
plt.show()

# Calculate and plot ROC curve with brighter colors
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})', color='blue', lw=2)
plt.plot([0, 1], [0, 1], linestyle='--', color='red')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()

# Display summary table of feature importance
print(feature_importance.head(20))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Ensure 'Flying_month' is properly formatted
df['Flying_month'] = pd.to_datetime(df['Flying_month'], format='%b-%y', errors='coerce')

# Convert categorical features to one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Traveller_type', 'Class', 'Route', 'Flying_month'])

# Define features (X) and target (y)
X = df_encoded.drop(columns=['Rating', 'Review_title', 'Review_content'])
y = df_encoded['Rating']

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Random Forest Regressor model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate and print the RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse}')

# Feature importance
feature_importance = pd.DataFrame(model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

# Plot the top 20 important features with brighter colors
plt.figure(figsize=(12, 8))
sns.barplot(x='importance', y=feature_importance.index[:20], data=feature_importance[:20], palette="viridis")
plt.title('Top 20 Feature Importances')
plt.show()

# Display summary table of feature importance
print(feature_importance.head(20))
```
